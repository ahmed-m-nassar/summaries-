## Range , Zip , Eumerate For loops
These are functions which can help in loops

#### 1- Range
Generates a sequence in a list 

```python
range(start, stop, step)
range(5) # 0 , 1 , 2 , 3 , 4
for x in range(len(list)):
  print(list[x])
```

#### 2- Zip 
used to group multiple data

```python
anguages = ['Java', 'Python', 'JavaScript']
versions = [14, 3, 6]
result = zip(languages, versions)	# Output: [('Java', 14), ('Python', 3), ('JavaScript', 6)]
l, v =  zip(*result)                # l = ['Java', 'Python', 'JavaScript']      , v = [14, 3, 6]
for l, n in zip(languages , versions):   # for loop using zip 
    print(l)
    print(n)
```

#### 3 - Enumerate 
Used to group indices with a list

```python
languages = ['Java', 'Python', 'JavaScript']
e = enumerate(languages)
print(list(e))     # [(0, 'Java'), (1, 'Python'), (2, 'JavaScript')]
for count, value in enumerate(languages):
     print(count, value)
```

## Functions and Documentation


```python
def population_density(population, land_area):
    """Calculate the population density of an area.

    INPUT:
    population: int. The population of that area
    land_area: int or float. This function is unit-agnostic, if you pass in values in terms
    of square km or square miles the function will return a density in those units.

    OUTPUT: 
    population_density: population / land_area. The population density of a particular area.
    """
    return population / land_area
```

## Exception handling
- The try block lets you test a block of code for errors.
- The except block lets you handle the error.
- The else block lets you execute code when there is no error.
- The finally block lets you execute code, regardless of the result of the try- and except blocks.

```python
#************************************************#
try:
    print("Hello")
except:
    print("Something went wrong")
else:
    print("Nothing went wrong")

#***********************************************#
try:
    print(x)
except:
    print("Something went wrong")
finally:
    print("The 'try except' is finished")

#************************************************#
#you can include a specific error type
try:
    c = ((a+b) // (a-b))
except ZeroDivisionError:
    print ("a/b result in 0")
else:
    print (c)
#************************************************#
```

#### 2 - Resampling & Time shifting & Rolling & Expanding


```python
#Resampling
############################################################################################################
df.resample(rule='A').mean()


def first_day(entry):
    """
    Returns the first instance of the period, regardless of sampling rate.
    """
    if len(entry):  # handles the case of missing data
        return entry[0]
    
df.resample(rule='A').apply(first_day)
############################################################################################################


#Shifting
############################################################################################################
df.shift(1)
df.shift(-1)
############################################################################################################


#Rolling
############################################################################################################
# 7 day rolling mean
df.rolling(window=7).mean()
############################################################################################################



#Expanding
############################################################################################################
df['Close'].expanding(min_periods=30).mean()
#############################################################################################################
```
